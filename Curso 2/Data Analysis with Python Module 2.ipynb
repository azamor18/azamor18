```python
from types import MethodDescriptorType
from types import MethodDescriptorType
#DATA NORMALIZATION

#Columna length, existen datos desde 150 a 250 
#columna height tiene rangos de 50 a 100

Se requiere normalizar para permitir comparasiones justas

#edad, tiene rangos de 0 a 100
#income, tiene rangos de 0 a 20, 000

si queremos hacer inferencias, como regresiones lineales, intrisecamente, se va a ver el resultado reflejado
pero no necesariamente significa que sea un predictor

Rangos de 0 a 1


Metodos para normalizar

1) Scaling = divide cada numero por el maximo valor del set de datos, el resultado queda en valores de 0 a 1
df['length'] = df['length']  / df['length'] .max()

2) min - max = toma cada valor (x_old) menos  el minumo valor del set de datos y lo divide entre el rango de los datos, resultado del valor entre 0 y 1
df['length']  = (df['length'] - df['length'] .min() / [df['length'] .max() - df['length'].min())

3) z-score = cada valor menos la media y luego la divides entre la std dev, los resultados esperadoes es cerano a 0 o entre +/- 3
df['length'] = (df['length'] -df['length'] .mean()/df['length'] .std())
```
